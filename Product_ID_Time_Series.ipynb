```python
import matplotlib.pyplot as plt
import pandas as pd

private_data = pd.read_csv("rollups_20170902_20180604.csv",index_col=None)
```

```python
private_data.pivot('public_product_id', 'Date', 'salesappdownloads')
```

```python
pivot = private_data.pivot('public_product_id', 'Date', 'salesappdownloads')
```

```python


test = pivot.T.reset_index()
test = pivot.T
test.index = pd.to_datetime(test.index)
print("They are {} apps".format(len(test.columns)))
```

```python
df_without_null = test[test.columns[~test.isnull().any()]]
```

```python
df_without_zero = df_without_null.loc[:, (df_without_null != 0.0).all()]
```

```python

df_5000app = df_without_zero.sample(5000, axis = 1)
df_5000app_120day = df_5000app.iloc[0:120,:]
df_5000app_120day.to_csv('df_5000app_120day.csv')
```

```python
df_5000app = df_without_zero.sample(1000, axis = 1)
df_5000app_120day = df_5000app.iloc[0:120,:]
df_5000app_120day.to_csv('df_1000app_120day.csv')
```
